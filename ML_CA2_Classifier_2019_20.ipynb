```python
#import libraries
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn.metrics import accuracy_score

columnsName = open("headers.txt")
text_reader = columnsName.read()

dataFrame = pd.read_csv('queries.csv',header = None, na_values = ["?", " ?", "? "])

#dataFrame.head()

dataFrame.columns = text_reader.split("\n")

dataFrame.head()
```

```python
dataFrame2 = pd.read_csv('trainingset.csv',header = None, na_values = ["?", " ?", "? "])

#dataFrame.head()

dataFrame2.columns = text_reader.split("\n")

#age, balance, day, campaign, pdays, previous, default, housing, loan - continous features
#job, martial, education, month, contact, dataFrame2.poutcome.value_counts()

#default, housing, loan could fit in either


dataFrame2.head()
```

```python
le = preprocessing.LabelEncoder()#
DictCollection = ["id"]
for key in dataFrame2.columns:
    if key not in DictCollection:
        ProcessedData[key] = dataFrame2[key]
        
dicthead = dataFrame2.head()
dictFrameTarget = dataFrame2['subscribed'].to_dict()
dictFrameFeatures = dataFrame2[['age',
                               'job',
                               'marital',
                               'education',
                               'default',
                               'balance ',
                               'housing',
                               'loan',
                               'contact',
                               'day',
                               'month',
                               'pdays',
                               'previous',
                               'poutcome']].to_dict()
for r in dictFrameTarget:
    if (dictFrameTarget[r] == "yes"):
        dictFrameTarget[r] = 1
    elif(dictFrameTarget[r] == "no"):
        dictFrameTarget[r] = 0
    
ProcessedData = {}

le = preprocessing.LabelEncoder()
DictCollection = ["id","subscribed"]
for key in dataFrame2.columns:
    if key not in DictCollection:
        ProcessedData[key] = le.fit_transform(dataFrame2[key])
        
        
df = pd.DataFrame(ProcessedData) 
#df.head()

train,test,train_labels,test_labels = train_test_split(df,
                                                       dictFrameTarget,
                                                       test_size = 0.33,
                                                       random_state=42)

gnb = GaussianNB()
model = gnb.fit(train, train_labels)

preds = gnb.predict(test)
#print(preds)

print(accuracy_score(test_labels, preds))
```
